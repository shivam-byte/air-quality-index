```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('/content/cleaned_AQI.csv')
```

```python
data
```

```python
data.drop('Unnamed: 0',axis = 1 ,inplace = True)
```

```python
X = data.drop('AQI',axis = 1)
y = data['AQI']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

```python
Dtree = DecisionTreeRegressor()
Dtree.fit(X_train,y_train)
```

```python
prediction = Dtree.predict(X_test)
```

```python
prediction
```

```python
mean_sq_error = mean_squared_error(y_test,prediction)
print('mean squared error:',mean_sq_error)
```

```python
params = {'max_depth':[3,5,6,8,10,15,18],'min_samples_split':[2,5,10],'min_samples_leaf':[1,2,5],'max_features': ['auto', 'sqrt', 'log2', None]}
```

```python
grid_search = GridSearchCV(estimator = Dtree,param_grid = params,cv = 5,n_jobs = 1,scoring = 'neg_mean_squared_error')
```

```python
grid_search.fit(X_train,y_train)
```

```python
print('best parameters :',grid_search.best_params_)
best_model = grid_search.best_estimator_
```

```python
prediction1 = best_model.predict(X_test)
```

```python
mse1 = mean_squared_error(y_test,prediction1)
print(mse1)
```
